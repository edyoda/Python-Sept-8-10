```python
def deco(func):
    
    def new_func(val1,val2):
        if type(val1) == type(val2):
            return func(val1,val2)
        else:
            return func(str(val1),str(val2))
    
    
    return new_func


@deco
def add(val1,val2):
    return val1+val2

print(add(10,20))

print(add("Python","Java"))

print(add("Python",100))


```

```python
import threading 
import time

def square(num1):
    print("Inside square")
    time.sleep(10)
    print(num1 * num1)
    
    
def multiply(num1,num2):
    print("Inside multiply")
    print(num1 * num2)
    
t1 = threading.Thread(target=square,args=(10,)) 
t2 = threading.Thread(target = multiply,args = (5,6))

t1.start()
t1.join(2)
t2.start()
```

```python
import threading 
import time

def square(num1):
    print("Inside square")
    time.sleep(10)
    print(num1 * num1)
    
    
def multiply(num1,num2):
    print("Inside multiply")
    print(num1 * num2)
    
t1 = threading.Thread(target=square,args=(10,),daemon=True) 
t2 = threading.Thread(target = multiply,args = (5,6))

t1.start()
t2.start()
```
